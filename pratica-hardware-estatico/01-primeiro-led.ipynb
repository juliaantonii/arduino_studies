# 1. Primeiro LED

Esta atividade marca o início prático dos estudos, focando na criação de um circuito fechado para acionar um componente de saída (LED).

O objetivo foi realizar a ligação de um LED de forma segura, utilizando uma fonte de energia do Arduino Uno R3. O projeto validou o conceito de polaridade e a necessidade de limitação de corrente através de um resistor.

# 2. Componentes Utilizados
* 01 Arduino Uno R3 (Fonte de 5V)
* 01 LED (Verde)
* 01 Resistor de 220 Ohms
* 01 Protoboard
* 05 Jumpers para conexão

# 3. O Circuito
Para que o LED acenda sem queimar, o circuito foi montado da seguinte forma:
* Saída de Energia: Conectado ao pino 5V do Arduino.
* Proteção: O Resistor (220 Ohms) foi colocado em série para limitar a corrente.
* Carga: O Anodo (+) do LED conectado ao resistor e o Catodo (-) conectado ao GND.
* Fechamento: O circuito retorna ao pino GND (Terra) do Arduino.
