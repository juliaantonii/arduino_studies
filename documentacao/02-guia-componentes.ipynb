# 1. Componentes Eletrônicos

A eletricidade se relaciona com o movimento de elétrons, a eletrônica tem a ver com aquilo que usamos para controlar o movimento dos elétrons, para isso, usamos os **Componentes Eletrônicos**.

## 1.1 Passivos e Proteção

* **Resistor (Resistência)**: O componente mais onipresente na eletrônica. Sua função é limitar o fluxo de elétrons.
    * Função: Reduzir a corrente elétrica e dividir tensões.
    * Unidade: Ohms (Ω).

![Resistor Color Codes](https://encrypted-tbn3.gstatic.com/licensed-image?q=tbn:ANd9GcRun5IsMetXEYPU-ELRLPXMXV_mLvxjStj2B_r4KqJtDit4S1rHiB7jZV7h9vALmZOq9UvMLE8mVXAld__V3tvR4dqXrbACu2525i0NEbVtGsMET20)

* **Capacitor**: O capacitor é um componente capaz de armazenar energia num campo elétrico. Ele "carrega" quando a eletricidade passa e "descarrega" quando a tensão cai.
    * Função: Estabilizar tensões, filtrar ruídos em sinais e criar temporização em circuitos analógicos.
    * Polaridade: Alguns possuem e outros não possuem. Necessário tomar cuidado, caso a polaridade seja invertida, há o risco dele explodir.
    * Unidade de Medida: Farad ($F$). Usamos valores muito pequenos como microfarads ($\mu F$), nanofarads ($nF$) ou picofarads ($pF$).

* **Diodo (Retificador)**: O diodo é um componente semicondutor que permite a passagem da corrente elétrica em apenas um sentido.
    * Função Principal: Proteger circuitos contra inversão de polaridade e converter corrente alternada (AC) em corrente contínua (DC).
    * Polaridade: Sim. Possui Anodo (positivo) e Catodo (negativo).
    * Identificação: No componente físico, o Catodo é marcado por uma faixa cinza ou preta em uma das extremidades.

## 1.2 Atuadores e Sinalização

* **LED (Diodo Emissor de Luz)**: Um semicondutor que converte energia elétrica em luz.
    * Função: Sinalização visual.
    * Polaridade: Sim. O Anodo (perna maior ou reta) é o positivo. O Catodo (perna menor ou com chanfro no bulbo) é o negativo (GND).

* **Buzzer (Transdutor Piezoelétrico)**: Um componente que converte sinais elétricos em ondas sonoras.
    * Função: Alertas sonoros ou melodias simples.
    * Tipos: * Ativo: Basta ligar no 5V e ele apita.
        * Passivo: Você precisa enviar uma frequência (PWM) usando a função tone() no Arduino para gerar som.

## 1.3 Interface e Sensores

* **Push Button (Interruptor Tátil)**: Um interruptor que fecha o circuito apenas quando pressionado.
    * Função: Entrada digital (comando do usuário).
    * Polaridade: Não possui, mas tem uma armadilha: ele possui 4 terminais. Internamente, os pinos laterais já são conectados aos pares.

* **Potenciômetro**: Um resistor cujo valor pode ser ajustado girando um eixo.
    * Função: Entrada analógica (controlar volume, brilho ou velocidade).
    * Pinos: Possui 3 pinos. Os das extremidades vão no VCC (5V) e GND. O pino central (cursor) vai em uma entrada analógica (A0-A5).

* **LDR (Resistor Dependente de Luz)**: Um componente cuja resistência muda de acordo com a intensidade de luz que incide sobre ele.
    * Função: Sensor de luminosidade.
    * Comportamento: Muita luz = Baixa resistência. Escuridão = Alta resistência.

# 2. Circuitos Eletrônicos

É fundamental compreender também os **Circuitos Eletrônicos**, que são um conjunto de componentes que formam o "caminho" onde os elétrons passarão. Todo circuito funcional precisa de:
* Fonte (Source): Fornece a energia (Ex: Bateria, porta USB do Arduino).
* Carga (Load): O que consome a energia para realizar um trabalho (Ex: LED, Buzzer).
* Condutores: O caminho por onde os elétrons fluem (Ex: Jumpers, trilhas da Protoboard).
* Controle (Opcional): O que interrompe ou altera o fluxo (Ex: Botão, Potenciômetro).

## 2.1 Topologias de Conexão

A forma como conectamos os componentes altera o comportamento da Tensão (V) e da Corrente (I).

**Série vs. Paralelo**:
* Série: A corrente tem apenas um caminho. Se um componente falhar, o circuito abre.
    * Corrente (I): É a mesma em todos os componentes.
    * Tensão (V): É dividida entre os componentes.
    * Ponto Crítico: Se um componente quebrar, o circuito inteiro para.

![Série](https://encrypted-tbn0.gstatic.com/licensed-image?q=tbn:ANd9GcRq0PTpWUgNbRrCghXNqQnKuUarjYc1cRBMWHzPYo4ZAILKvV2LY8K9oT6f8TILftXfmRv2M7Q_87-eS4kezvfCFnrf40EeFVS-7E7UMCWiVxZ9rqs)

* Paralelo: A corrente se divide em múltiplos caminhos. Se um LED apagar, os outros continuam acesos.
    * Corrente (I): Se divide entre os caminhos.
    * Tensão (V): É a mesma para todos os componentes.
    * Vantagem: Se um componente falhar, os outros continuam funcionando normalmente.
