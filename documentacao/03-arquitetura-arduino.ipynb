# 1. Arquitetura do Arduino
O Arduino Uno é baseado no microcontrolador ATmega328P. Ele atua como o cérebro do sistema, recebendo informações dos sensores (entradas), processando-as conforme o código e enviando comandos para os atuadores (saídas).

## 1.1 Mapeamento da Placa (Pinout)
Entender a função de cada pino é crucial para não causar danos ao hardware.
* **Pinos Digitais (0 a 13)**: Podem assumir apenas dois estados: HIGH (5V) ou LOW (0V).
    * Nota: Os pinos com o símbolo ~ (3, 5, 6, 9, 10 e 11) suportam PWM (simulação de saída analógica).
* **Pinos Analógicos (A0 a A5)**: Usados para ler sensores que variam continuamente (ex: LDR ou Potenciômetro). Eles convertem a tensão (0V a 5V) em um valor digital de 0 a 1023.
* **Pinos de Energia**:
    * 5V e 3.3V: Saídas de alimentação para os componentes.
    * GND (Ground): O referencial 0V do circuito (Terra).
    * VIN: Entrada para alimentação externa (bateria).

## 1.2 O Ciclo de Execução
Diferente de um script Python linear, o Arduino opera em um loop infinito. A estrutura básica de qualquer programa (`sketch`) é:
* void `setup()`: Executado apenas uma vez quando a placa é ligada. Usado para configurar os pinos (pinMode).
* void `loop()`: Executado repetidamente enquanto a placa tiver energia. É onde a lógica principal reside.

## 1.3 O Arduino se comunica com o computador através do cabo USB via protocolo Serial.

Baud Rate: Velocidade da comunicação (comumente 9600 bps).

Monitor Serial: Ferramenta da IDE que permite "imprimir" valores para debug.

```python
# Cálculo de resolução do Arduino (10 bits)
v_ref = 5.0 # Tensão de referência
bits = 1024 # 2^10

def ler_tensao(valor_adc):
    return (valor_adc * v_ref) / bits

print(f"Um valor ADC de 512 equivale a {ler_tensao(512)}V")
```

## 1.4 Desenvolvimento
Para garantir a segurança dos componentes e a fidelidade dos testes, o fluxo de estudo é dividido em duas etapas:

* **Simulação ([Tinkercad](https://www.tinkercad.com/))**: Utilizado para prototipagem inicial, validação de circuitos e testes de lógica sem risco de danos ao hardware.
    * Vantagem: Testar componentes que ainda não possuo fisicamente e validar cálculos de resistores.

* **Hardware Real ([Arduino Uno R3](https://docs.arduino.cc/hardware/uno-rev3/))**: Implementação física dos circuitos validados.
    * Vantagem: Lidar com variáveis reais como ruído elétrico, conexões físicas em protoboard e comportamento real de sensores.

> Nota: As pinagens e comportamentos documentados aqui são 100% compatíveis entre o ambiente virtual e o físico, respeitando os limites de corrente do microcontrolador ATmega328P.
