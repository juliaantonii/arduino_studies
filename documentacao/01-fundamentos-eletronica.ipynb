# 1. Grandezas Elétricas Fundamentais

As quatro grandezas elétricas fundamentais são tensão (Volts, V), corrente (Ampères, A), resistência (Ohms, Ω) e potência (Watts, W). Estas descrevem o funcionamento dos circuitos: a tensão impulsiona, a corrente flui, a resistência opõe-se e a potência é a taxa de consumo/transferência de energia.

Devemos lembrar das características da natureza: quando tenho dois objetos com temperaturas diferentes, sabemos que o objeto com temperatura maior irá compartilhar a temperatura com o outro, até que este atinjam a mesma temperatura, ou seja, há a busca de equilíbrio.

Seguindo a lógica, temos a composição da matéria. Toda matéria é formada por átomos e estes são compostos pelo núcleo (prótons e neutrons) e, orbitando o núcleo, temos os elétrons. Onde:
* **Prótons**: Possuem carga elétrica positiva.
* **Neutrons**: Não possuem carga elétrica.
* **Elétrons**: Possuem carga elétrica negativa.

## 1.1 Tensão

Imaginando essa diferença de quantidade de elétrons entre objetos, conseguimos definir a primeira grandeza, a **Tensão**. Dizemos que a tensão elétrica é quando tenho uma diferença de potencial elétrico entre dois pontos/objetos/corpos, utilizando como unidade de medida o **Volts / V**. Um exemplo prático seria uma tomada elétrica residencial, que no padrão brasileiro utiliza de 3 pinos, possuindo:
* **Fase (L)**: Fio energizado, onde está presente a tensão elétrica.
* **Neutro (N)**: Fio de retorno da corrente elétrica, com potencial próximo de 0V.
* **Terra (PE)**: Fio de proteção conectado ao sistema de aterramento elétrico, utilizado para segurança contra choques elétricos.

## 1.2 Corrente

Tendo a diferença de potencial elétrico, conseguimos compreender o conceito de **Corrente**. Compreendemos como o fluxo ordenado de cargas elétricas (geralmente elétrons) que percorre um condutor quando existe uma diferença de potencial, utilizando como unidade de medida o **Ampère / A**.

## 1.3 Resistência

É a propriedade física que materiais condutores possuem de opor-se à passagem da corrente elétrica, dificultando o fluxo de elétrons devido a colisões atômicas, sendo capaz de elevar a temperatura no meio em que ela esta circulando, utilizando como unidade de medida o **Ohms / Ω**. Podemos afirmar também que é uma característica do componente **resistor**.

Um exemplo muito comum dessa propriedade é o funcionamento de um Chuveiro elétrico. Dentro do chuveiro existe um fio metálico chamado Resistência de chuveiro (ou elemento resistivo). Quando a Corrente elétrica passa por esse fio:
* 1. Os elétrons se movimentam pelo material.
* 2. Durante esse movimento ocorrem colisões entre elétrons e átomos do material.
* 3. Essas colisões transformam parte da energia elétrica em calor.

Esse calor aquece a água que passa pelo chuveiro, permitindo o banho quente.

## 1.4 Potência

Esta mede a quantidade de trabalho realizado em determinado intervalo de tempo, ou seja, é a taxa de variação da energia, utilizando como unidade de medida o **Watts / W**. Sabendo que "trabalho" refere-se à energia elétrica consumida ou transformada por um dispositivo ao longo do tempo.

# 2. A Lei de Ohm

## 2.1 Primeira Lei de Ohm

A primeira lei determina a relação linear entre tensão, corrente e resistência. Os resistores que obedecem a essa lei são chamados de ôhmicos ou lineares.
* Fórmula: U = R x I
* Variações:
    * Resistência: R = U / I
    * Corrente: I = U / R

## 2.2 Segunda Lei de Ohm
Enquanto a primeira lei foca no circuito, a segunda lei de Ohm relaciona a resistência de um condutor às suas características físicas: material, comprimento e área da secção transversal.
* Fórmula: R = p L / A
* Variações:
    * (rô): Resistividade do material.
    * Comprimento do condutor.
    * Área da secção transversal (espessura).
